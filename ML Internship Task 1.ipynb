# Predicting if the user will buy the phone or not

```python
#Importing the libraries
import numpy as np
import pandas as pd
from sklearn.preprocessing import Imputer, LabelEncoder, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

## Reading the data
 **Here we use the pandas.read_csv() to read the data from the computer into our program and store it in a variable named as dataset**
 

```python
#Importing the dataset
dataset = pd.read_csv("dataset.csv")
```

## Making the dataframe
 **In order to alter the dataset we make the dataframe of the features that are important to us ,
    we select the particular columns/features from the dataset by using the iloc().**
    

```python
#Making the dataframe for the Independent Variables/Features
x = dataset.iloc[:,[4,5,6,7,8,9,10,12,13,14,15,18,20,22,23,24,25,26,29,32,34,35,40,42,44,45,46,48,50,51,52,53,54,55,56,57,65,67,68,69,70,74,80,81,82,83,84,85,86,87,89,92,93,96,97]]
```

## Preprocessing
 **The wireless charging feature conatains a lot of empty spaces so by using this loop we fill in the empty spaces which would further help us in encoding of the data**

```python
for i in range(len(x["Wireless Charging"])):
    if x["Wireless Charging"][i]!="yes":
        x["Wireless Charging"][i]="no"
```

**Our dataframe contains a lot of empty spaces or white spaces so in order to fill them with some string value we use the replace() where in it replces all the empty spaces by nan, also as our datser is quite huge with around more than 200 rows we bring it down to 10 rows using head()**

```python
x = x.replace(np.nan,'',regex=True)
x.head(10)
```

## Encoding
**Here we label encode our datframe and convert all the string values into numeric values using the le.fit_transform ().
the loop runs for every iteration and in each iteration it transforms the feature during that iteration**

```python
for i in range(55):
    le = LabelEncoder()
    x.iloc[:,i] = le.fit_transform(x.iloc[:,i])
```

```python
x.head(10)
```

**Imputer is used to convert the empty spaces which have been replaced by strings into some numeric values , these values can be the mean ,median ,or most frequent terms in that column or feature 
we use imputer because in real life scenarios as well there are situaions where data is not present and which is valuable so by using this the original data is not lost and the model too is not compromised
in our case we use the mean method to replace the empty spaces.**

```python
imputer = Imputer(missing_values = 0, strategy = "mean", axis = 0)
imputer = imputer.fit(x.iloc[:,0:])
x.iloc[:,0:] = imputer.transform(x.iloc[:,0:])
```

**So the preprocessed data is now again converted to csv format**

```python
x.to_csv("x_data.csv")
x.head(10)
```

**We use one hot ecoding to convert the categorical data as our algorithm cannot work with categorical data so we assign some value to that data 
and use it for further process, here onehot.fit_transform() is used for coding and it stores it in the variable x which is our dataset.**

```python
onehot = OneHotEncoder(categorical_features = [6])
x = onehot.fit_transform(x).toarray()
```

```python
x_pd = pd.DataFrame(x)
x_pd.head(10)
```

```python
x_pd.to_csv("x_final.csv")
```

## spliting the data 
 **we remove the test part from the training part using the iloc()**

```python
y = dataset.iloc[:,-1].values
```

### setting the parameters for selection
 ** Here we set the selection papameter as 3.5 so in the loop it gives the testing set as if the value is 3.5 then reppaces it with one or else replace it with 0**

```python
for i in range(len(y)):
    if y[i]>=3.5:
        y[i]=1
    else:
        y[i]=0
```

### Training and Spliting 
 **We use the test,train and split function of the sklearn library to train  test and split our data where 80% is our training data and the remaining 20% is our testing data**

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.2, random_state = 10)
```

## Building the model 
**Over here we use the decission tree classifier as our model to classify and predict the values** 

```python
dt=DecisionTreeClassifier(criterion='entropy',random_state=20)
```

```python
dt.fit(xtrain,ytrain)
```

```python
ypred = dt.predict(xtest)
```

### Testing the accuracy of our model 

```python
print(accuracy_score(ytest,ypred))
```
